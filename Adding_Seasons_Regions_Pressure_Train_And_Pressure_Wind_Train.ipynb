```python
import cv2
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
pressure_training_data_path = "/content/drive/MyDrive/First2YearRawAndNormData/CESMZ200/MinMaxNormWithLabel"
CL_path = pressure_training_data_path + "/CL"
COH_path = pressure_training_data_path + "/COH"
COL_path = pressure_training_data_path + "/COL"
NROI_path = pressure_training_data_path + "/NROI"
```

```python
pressure_CL_files = [f for f in os.listdir(CL_path) if f.endswith('.csv')]
pressure_COH_files = [f for f in os.listdir(COH_path) if f.endswith('.csv')]
pressure_COL_files = [f for f in os.listdir(COL_path) if f.endswith('.csv')]
pressure_NROI_files = [f for f in os.listdir(NROI_path) if f.endswith('.csv')]
```

```python
pressure_array = []
for file in pressure_CL_files:
    dataframe = pd.read_csv(os.path.join(CL_path, file), header=None)
    dataframe = dataframe.to_numpy().flatten()
    name = os.path.splitext(file)[0]
    part = name.split("_")
    day = int(part[0][3:])
    dataframe = np.append(dataframe, name)
    dataframe = np.append(dataframe, day)
    pressure_array.append(dataframe)

for file in pressure_COH_files:
    dataframe = pd.read_csv(os.path.join(COH_path, file), header=None)
    dataframe = dataframe.to_numpy().flatten()
    name = os.path.splitext(file)[0]
    part = name.split("_")
    day = int(part[0][3:])
    dataframe = np.append(dataframe, name)
    dataframe = np.append(dataframe, day)
    pressure_array.append(dataframe)

for file in pressure_COL_files:
    dataframe = pd.read_csv(os.path.join(COL_path, file), header=None)
    dataframe = dataframe.to_numpy().flatten()
    name = os.path.splitext(file)[0]
    part = name.split("_")
    day = int(part[0][3:])
    dataframe = np.append(dataframe, name)
    dataframe = np.append(dataframe, day)
    pressure_array.append(dataframe)

for file in pressure_NROI_files:
    dataframe = pd.read_csv(os.path.join(NROI_path, file), header=None)
    dataframe = dataframe.to_numpy().flatten()
    name = os.path.splitext(file)[0]
    part = name.split("_")
    day = int(part[0][3:])
    dataframe = np.append(dataframe, name)
    dataframe = np.append(dataframe, day)
    pressure_array.append(dataframe)
```

```python
Y = []
for i in range(len(pressure_CL_files)):
  Y.append(0)

for i in range(len(pressure_COH_files)):
  Y.append(1)

for i in range(len(pressure_COL_files)):
  Y.append(2)

for i in range(len(pressure_NROI_files)):
  Y.append(3)
```

```python
wind_training_data_path = "/content/drive/MyDrive/First2YearRawAndNormData/CESMU200/MinMaxNormWithLabel"
wind_CL_path = wind_training_data_path + "/CL"
wind_COH_path = wind_training_data_path + "/COH"
wind_COL_path = wind_training_data_path + "/COL"
wind_NROI_path = wind_training_data_path + "/NROI"
```

```python
CL_wind_files = [f for f in os.listdir(wind_CL_path) if f.endswith(".csv")]
COH_wind_files = [f for f in os.listdir(wind_COH_path) if f.endswith(".csv")]
COL_wind_files = [f for f in os.listdir(wind_COL_path) if f.endswith(".csv")]
NROI_wind_files = [f for f in os.listdir(wind_NROI_path) if f.endswith(".csv")]
```

```python
wind_array = []
for file in CL_wind_files:
    dataframe = pd.read_csv(os.path.join(CL_path, file), header=None)
    dataframe = dataframe.to_numpy().flatten()
    wind_array.append(dataframe)

for file in COH_wind_files:
    dataframe = pd.read_csv(os.path.join(COH_path, file), header=None)
    dataframe = dataframe.to_numpy().flatten()
    wind_array.append(dataframe)

for file in COL_wind_files:
    dataframe = pd.read_csv(os.path.join(COL_path, file), header=None)
    dataframe = dataframe.to_numpy().flatten()
    wind_array.append(dataframe)

for file in NROI_wind_files:
    dataframe = pd.read_csv(os.path.join(NROI_path, file), header=None)
    dataframe = dataframe.to_numpy().flatten()
    wind_array.append(dataframe)
```

```python
pressure_training_df = pd.DataFrame(pressure_array)
pressure_training_df.rename(columns = {225: "Name", 226: "Day"}, inplace = True)
wind_training_df = pd.DataFrame(wind_array)
```

```python
#Normalize the data
scaler = MinMaxScaler()
scaled_pressure = scaler.fit_transform(pressure_training_df.drop(columns = ["Name", "Day"]))
scaled_wind = scaler.fit_transform(wind_training_df)
```

```python
#Combine the data
merging_data = np.stack((scaled_pressure, scaled_wind), axis = 2)

#Reshape the data
(h, w, c) = merging_data.shape
reshaped_df = merging_data.reshape(h, w*c)
```

```python
wind_pressure_training_data = pd.DataFrame(reshaped_df)
wind_pressure_training_data["Name"] = pressure_training_df["Name"]
wind_pressure_training_data["Day"] = pressure_training_df["Day"]
wind_pressure_training_data
```

# **Adding Regions and Seasons for Pressure Training Data**

```python
# Get the index of the "224" column
index_224 = pressure_training_df.columns.get_loc(224)
```

```python
image = cv2.imread("/content/drive/MyDrive/Pressure_raw_Image/day1.png")

#Get the dimension of the image
height, width, _ = image.shape

#calculate the width and height for each region
region_width = width // 3
region_height = height // 3

regions0 = []
regions1 = []
regions2 = []
regions3 = []
divided_regions = ["0000", "0010", "0110", "1000", "1010", "1110", "1001", "1011", "1111"]
items = np.array(pressure_training_df["Name"])
for item in items:
    part = item.split("_")
    x = int(part[3])
    y = int(part[4])

    row_index = y // region_height
    col_index = x // region_width

    region_number = row_index * 3 + col_index
    regions0.append(int(divided_regions[region_number][0]))
    regions1.append(int(divided_regions[region_number][1]))
    regions2.append(int(divided_regions[region_number][2]))
    regions3.append(int(divided_regions[region_number][3]))

new_columns = pd.DataFrame({
    225: regions0,
    226: regions1,
    227: regions2,
    228: regions3
})
pressure_df = pd.concat([pressure_training_df.iloc[:, :index_224 + 1], new_columns, pressure_training_df.iloc[:, index_224 + 1:]], axis=1)
```

```python
insert_index = pressure_df.columns.get_loc(228) + 1
```

```python
pressure_df['Day'] = pressure_df['Day'].astype(int)
# Define a function to map day values to seasons
def map_season(day):
    year = (day - 1) // 365 + 1  # Calculate the year based on the day value
    day_within_year = (day - 1) % 365 + 1  # Calculate the day within the year

    if 1 <= day_within_year <= 90:          # Spring: Month 1-3
        return 1
    elif 91 <= day_within_year <= 180:      # Summer: Month 4-6
        return 2
    elif 181 <= day_within_year <= 270:     # Fall: Month 7-9
        return 3
    else:                                  # Winter: Month 10-12
        return 4

# Apply the function to the 'Day' column to create the 'season' column
pressure_df['season'] = pressure_df ['Day'].apply(map_season)
pressure_df.insert(insert_index, 'season', pressure_df.pop('season'))
```

```python
pressure_df['Class'] = Y
pressure_df
```
