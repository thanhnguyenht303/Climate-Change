```python
import os
import cv2
import pandas as pd
import numpy as np
```

```python
Pressure_raw_folder = "/content/drive/MyDrive/Pressure_raw_Image"
ROI_Pressure_folder = "/content/drive/MyDrive/ROI_Pressure_Image"

SOM_Pressure_folder = "/content/drive/MyDrive/4th_5th_SOM_Categorical_Pressure_ROI_Image"
SOM_Pressure_Wind_folder = "/content/drive/MyDrive/4th_5th_SOM_Categorical_Wind_Pressure_ROI_Image"
SOM_Pressure_Location_Season_folder = "/content/drive/MyDrive/4th_5th_SOM_Categorical_Pressure_Location_Season_ROI_Image"
SOM_Pressure_Wind_Location_Season_folder = "/content/drive/MyDrive/4th_5th_SOM_Categorical_Wind_Pressure_Location_Season_ROI_Image"

CNN_Pressure_folder = "/content/drive/MyDrive/4th_5th_CNN_Categorical_Pressure_ROI_Image"
CNN_Pressure_Wind_folder = "/content/drive/MyDrive/4th_5th_CNN_Categorical_Wind_Pressure_ROI_Image"
CNN_Pressure_Location_Season_folder = "/content/drive/MyDrive/4th_5th_CNN_Categorical_Pressure_Location_Season_ROI_Image"
CNN_Pressure_Wind_Location_Season_folder = "/content/drive/MyDrive/4th_5th_CNN_Categorical_Wind_Pressure_Location_Season_ROI_Image"

```

```python
Output_folder = "/content/drive/MyDrive/4th_5th_Merging_Image"

# os.makedirs(Output_folder)
```

```python
for file in os.listdir(SOM_Pressure_folder):
    path3 = os.path.join(SOM_Pressure_folder, file)

    if os.path.exists(os.path.join(Pressure_raw_folder, file)) and os.path.exists(os.path.join(ROI_Pressure_folder, file)) and os.path.exists(os.path.join(SOM_Pressure_Wind_folder, file)) and os.path.exists(os.path.join(SOM_Pressure_Location_Season_folder, file)) and os.path.exists(os.path.join(SOM_Pressure_Wind_Location_Season_folder, file)) and os.path.exists(os.path.join(CNN_Pressure_folder, file)) and os.path.exists(os.path.join(CNN_Pressure_Wind_folder, file)) and os.path.exists(os.path.join(CNN_Pressure_Location_Season_folder, file)) and os.path.exists(os.path.join(CNN_Pressure_Wind_Location_Season_folder, file)):

        path1 = os.path.join(Pressure_raw_folder, file)
        path2 = os.path.join(ROI_Pressure_folder, file)
        path4 = os.path.join(SOM_Pressure_Wind_folder, file)
        path5 = os.path.join(SOM_Pressure_Location_Season_folder, file)
        path6 = os.path.join(SOM_Pressure_Wind_Location_Season_folder, file)
        path7 = os.path.join(CNN_Pressure_folder, file)
        path8 = os.path.join(CNN_Pressure_Wind_folder, file)
        path9 = os.path.join(CNN_Pressure_Location_Season_folder, file)
        path10 = os.path.join(CNN_Pressure_Wind_Location_Season_folder, file)


        image1 = cv2.imread(path1)
        image2 = cv2.imread(path2)
        image3 = cv2.imread(path3)
        image4 = cv2.imread(path4)
        image5 = cv2.imread(path5)
        image6 = cv2.imread(path6)
        image7 = cv2.imread(path7)
        image8 = cv2.imread(path8)
        image9 = cv2.imread(path9)
        image10 = cv2.imread(path10)

        above_img = np.concatenate((image1, image2, image3, image7, image4, image8), axis=1)
        under_img = np.concatenate((image5, image9, image6, image10), axis=1)

        height1, width1, _ = above_img.shape
        height2, width2, _ = under_img.shape

        if width1 >= width2:
            white_img = np.ones((height2, width1, 3), dtype=np.uint8) * 255
            white_img[:, 0:0+width2, :] = under_img
            merged_image = np.concatenate((above_img, white_img), axis=0)
        else:
            white_img = np.ones((height1, width2, 3), dtype = np.uint8) * 255
            white_img[:, 0:0+width1, :] = above_img
            merged_image = np.concatenate((white_img, under_img), axis=0)

    output_file_path = os.path.join(Output_folder, file)
    cv2.imwrite(output_file_path, merged_image)
```

```python
file_count = len([f for f in os.listdir(Output_folder) if os.path.isfile(os.path.join(Output_folder, f))])

print(file_count)
```
