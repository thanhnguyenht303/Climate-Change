```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image = Image.open("/content/day1.png")
width, height = image.size
print(width, height)
```

```python
plt.imshow(image, cmap = "gray")
plt.show()
```

```python
n_width, n_height = width * 2, height * 2
n_image = image.resize((n_width, n_height))
```

```python
plt.imshow(n_image)
plt.show()
```
