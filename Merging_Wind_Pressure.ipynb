```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
pressure_path = "/content/drive/MyDrive/Pressure Region of Interest"
wind_path = "/content/drive/MyDrive/Wind_ROI_points"
```

```python
pressure_arr = []
for file in os.listdir(pressure_path):
    file_path = os.path.join(pressure_path, file)
    df = pd.read_csv(file_path, header = None)
    flatten_df = df.to_numpy().flatten()
    name = os.path.splitext(file)[0]
    part = name.split("_")
    day = int(part[0])
    flatten_df = np.append(flatten_df, name)
    flatten_df = np.append(flatten_df, day)
    pressure_arr.append(flatten_df)
```

```python
pressure_data = pd.DataFrame(pressure_arr)
pressure_data.rename(columns = {225: "Name", 226: "Day"}, inplace = True)
pressure_data
```

```python
wind_arr = []
for file in os.listdir(wind_path):
    file_path = os.path.join(wind_path, file)
    df = pd.read_csv(file_path, header = None)
    flatten_df = df.to_numpy().flatten()
    name = os.path.splitext(file)[0]
    part = name.split("_")
    day = int(part[0])
    flatten_df = np.append(flatten_df, name)
    flatten_df = np.append(flatten_df, day)
    wind_arr.append(flatten_df)
```

```python
wind_data = pd.DataFrame(wind_arr)
wind_data.rename(columns = {225: "Name", 226: "Day"}, inplace = True)
wind_data
```

```python
#Normalize the data
scaler = MinMaxScaler()
scaled_pressure = scaler.fit_transform(pressure_data.drop(columns = ["Name", "Day"]))
scaled_wind = scaler.fit_transform(wind_data.drop(columns = ["Name", "Day"]))
```

```python
#Combine the data
merging_data = np.stack((scaled_pressure, scaled_wind), axis = 2)

#Reshape the data
(h, w, c) = merging_data.shape
reshaped_df = merging_data.reshape(h, w*c)
```

```python
data = pd.DataFrame(reshaped_df)
data["Name"] = pressure_data["Name"]
data["Day"] = pressure_data["Day"]
data
```

```python
folder_name = "Wind_Pressure_ROI_points"
folder_path = "/content/drive/MyDrive/" + folder_name

os.makedirs(folder_path)
```

```python
data.to_csv(folder_path + "/Wind_Pressure_ROI.csv", index = False)
```
