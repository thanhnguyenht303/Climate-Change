```python
import os
import cv2
import pandas as pd
import numpy as np
```

```python
Pressure_raw_folder = "/content/drive/MyDrive/Pressure_raw_Image"
ROI_Pressure_folder = "/content/drive/MyDrive/ROI_Pressure_Image"
SOM_Categorical_ROI_Pressure_folder_2year = "/content/drive/MyDrive/10_years_SOM_Categorical_Pressure_ROI_Image"
SOM_Categorical_ROI_Pressure_folder_3year = "/content/drive/MyDrive/10_years_SOM_Categorical_Pressure_ROI_Image(3years)"
```

```python
output_folder = "/content/drive/MyDrive/10years_Merging_Image"
os.makedirs(output_folder)
```

```python
for file in os.listdir(SOM_Categorical_ROI_Pressure_folder_2year):
    file_path3 = os.path.join(SOM_Categorical_ROI_Pressure_folder_2year, file)



    if os.path.exists(os.path.join(ROI_Pressure_folder, file)) and os.path.exists(os.path.join(Pressure_raw_folder, file)):
        file_path2 = os.path.join(ROI_Pressure_folder, file)
        file_path1 = os.path.join(Pressure_raw_folder, file)
        file_path4 = os.path.join(SOM_Categorical_ROI_Pressure_folder_3year, file)

        image1 = cv2.imread(file_path1)
        image2 = cv2.imread(file_path2)
        image3 = cv2.imread(file_path3)
        image4 = cv2.imread(file_path4)

        height, width, channel = image1.shape

        white_img = np.full((height, width, channel), 255, dtype=np.uint8)
        image1 = np.concatenate((image1, white_img), axis = 0)
        image2 = np.concatenate((image2, white_img), axis = 0)
        merged_img = np.concatenate((image1, image2, image3, image4), axis = 1)
    else:
        merged_img = cv2.imread(file_path1)
        merged_img = np.concatenate((merged_img, merged_img, merged_img, merged_img), axis = 1)

    output_file_path = os.path.join(output_folder, file)

    cv2.imwrite(output_file_path, merged_img)
```
