```python
import os
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Pressure_raw_folder = "/content/drive/MyDrive/Pressure_raw_Image"
ROI_Pressure_folder = "/content/drive/MyDrive/ROI_Pressure_Image"
Third_year_SOM_Categorical_ROI_folder = "/content/drive/MyDrive/3rd_SOM_Categorical_ROI_Pressure_Image"
Third_year_CNN_Categorical_ROI_folder = "/content/drive/MyDrive/3rd_CNN_Categorical_ROI_Pressure_Image"
Output_folder = "/content/drive/MyDrive/3rd_Merging_Image"
```

```python
for file in os.listdir(Third_year_CNN_Categorical_ROI_folder):
    file_path1 = os.path.join(Third_year_CNN_Categorical_ROI_folder, file)

    if os.path.exists(os.path.join(ROI_Pressure_folder, file)) and os.path.exists(os.path.join(Third_year_SOM_Categorical_ROI_folder, file)) and os.path.exists(os.path.join(Pressure_raw_folder, file)):
        file_path2 = os.path.join(ROI_Pressure_folder, file)
        file_path3 = os.path.join(Third_year_SOM_Categorical_ROI_folder, file)
        file_path4 = os.path.join(Pressure_raw_folder, file)

        image1 = cv2.imread(file_path1)
        image2 = cv2.imread(file_path2)
        image3 = cv2.imread(file_path3)
        image4 = cv2.imread(file_path4)

        merged_img = np.concatenate((image4, image2, image3, image1), axis = 1)

    output_file_path = os.path.join(Output_folder, file)

    cv2.imwrite(output_file_path, merged_img)
```
