```python
import os
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Pressure_raw_folder = "/content/drive/MyDrive/Pressure_raw_Image"
ROI_Pressure_folder = "/content/drive/MyDrive/ROI_Pressure_Image"
Categorical_ROI_Pressure_folder = "/content/drive/MyDrive/Categorical_ROI_Pressure_Image"
Output_folder = "/content/drive/MyDrive/Output_Image"
```

```python
for file in os.listdir(Pressure_raw_folder):
    file_path1 = os.path.join(Pressure_raw_folder, file)

    if os.path.exists(os.path.join(ROI_Pressure_folder, file)) and os.path.exists(os.path.join(Categorical_ROI_Pressure_folder, file)):
        file_path2 = os.path.join(ROI_Pressure_folder, file)
        file_path3 = os.path.join(Categorical_ROI_Pressure_folder, file)

        image1 = cv2.imread(file_path1)
        image2 = cv2.imread(file_path2)
        image3 = cv2.imread(file_path3)

        merged_img = np.concatenate((image1, image2, image3), axis = 1)
    else:
        merged_img = cv2.imread(file_path1)
        merged_img = np.concatenate((merged_img, merged_img, merged_img), axis = 1)

    output_file_path = os.path.join(Output_folder, file)

    cv2.imwrite(output_file_path, merged_img)
```
