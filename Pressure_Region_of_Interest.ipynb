```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_table("/content/InterestImageCenter.txt", sep = "_", header = None)
df.columns = ["Day", "Cluster", "ConnectedComponent", "X", "Y"]
df.info()
```

```python
pressure_raw = "/content/drive/MyDrive/Pressure_rawdata"
hash_table = {}
for file in os.listdir(pressure_raw):
  if file.endswith(".csv"):
    day_number = int(file[3:-4])
    hash_table[day_number] = file
```

```python
folder_name = "Pressure Region of Interest"
folder_path = "/content/drive/MyDrive/" + folder_name
os.makedirs(folder_path)
```

```python
path = "/content/drive/MyDrive/Pressure Region of Interest/"
rec_size = 15
for i in range(df.shape[0]):
  day = df["Day"][i]
  c = df["Cluster"][i]
  cp = df["ConnectedComponent"][i]
  cx = df["X"][i]
  cy = df["Y"][i]
  file_path = os.path.join(pressure_raw, hash_table[day])
  data = pd.read_csv(file_path, header = None)
  start_row = cy - rec_size//2
  start_col = cx - rec_size//2
  rectangle = data.iloc[start_row : start_row + rec_size, start_col : start_col + rec_size]
  rectangle.to_csv(path + str(day) + "_" + c + "_" + cp + "_" + str(cx) + "_" + str(cy) + ".csv", index = False, header = None)
```
