```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
pressure_raw = "/content/drive/MyDrive/Pressure_rawdata/"
image_list = []
for i in range(1, 3651):
  data = pd.read_csv(pressure_raw + "day" + str(i) + ".csv", header = None)
  A = data.to_numpy()
  res_sum = A // 100 * 100
  image_list.append(res_sum)
```

```python
folder_name = "Pressure_raw_Image"
folder_path = "/content/drive/MyDrive/" + folder_name

os.makedirs(folder_path)
```

```python
for i in range(len(image_list)):
  plt.imsave(folder_path + "/day" + str(i + 1) + ".png", image_list[i], cmap = "gray")
```
