```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
import pandas as pd
import os
```

```python
df = pd.read_table("/content/InterestImageCenter.txt", sep = "_", header = None)
df.columns = ["Day", "Cluster", "ConnectedComponent", "X", "Y"]
df.info()
```

```python
hash_table = {}
length = df.shape[0]
for i in range(length):
  day = df["Day"][i]
  x = df["X"][i]
  y = df["Y"][i]
  if day in hash_table:
    hash_table[day].append([x, y])
  else:
    hash_table[day] = [[x, y]]

# print(hash_table)
```

```python
folder_name = "ROI_Pressure_Image"
folder_path = "/content/drive/MyDrive/" + folder_name

os.makedirs(folder_path)
```

```python
image_list = []
image_path = "/content/drive/MyDrive/Pressure_raw_Image"
for file in os.listdir(image_path):
    file_path = os.path.join(image_path, file)

    final_image = cv2.imread(file_path)

    file_name = os.path.splitext(file)[0]
    number = int(file_name[3:])

    if number in hash_table:
        for point in hash_table[number]:
            x = point[0]
            y = point[1]
            size = 2
            color = (0, 0, 255)

            vertices = np.array([(x, y - size), (x - size, y), (x, y + size), (x + size, y)])

            cv2.drawContours(final_image, [vertices.reshape((-1, 1, 2))], -1, color, thickness = cv2.FILLED)

        cv2.imwrite(folder_path + "/day" + str(number) + ".png", final_image)


```
