```python
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import cv2
```

```python
data_path = "/content/drive/MyDrive/Pressure Region of Interest"
array = []
for file in os.listdir(data_path):
    file_path = os.path.join(data_path, file)
    df = pd.read_csv(file_path, header = None)
    flatten_df = df.to_numpy().flatten()
    name = os.path.splitext(file)[0]
    part = name.split("_")
    day = int(part[0])
    flatten_df = np.append(flatten_df, name)
    flatten_df = np.append(flatten_df, day)
    array.append(flatten_df)
```

```python
pressure_df = pd.DataFrame(array)
pressure_df.rename(columns={225: "Name", 226: "Day"}, inplace=True)
pressure_df
```

```python
# Normalize the first 225 columns
scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(pressure_df.iloc[:, :225])

# Create a DataFrame from the normalized values
normalized_df = pd.DataFrame(data_normalized, columns=pressure_df.columns[:225])

# Add the normalized columns back to the original DataFrame
pressure_df = pd.concat([normalized_df, pressure_df.iloc[:, 225:]], axis=1)
```

```python
pressure_df
```

```python
# Get the index of the "224" column
index_224 = pressure_df.columns.get_loc(224)
```

```python
image = cv2.imread("/content/drive/MyDrive/Pressure_raw_Image/day1.png")

#Get the dimension of the image
height, width, _ = image.shape

#calculate the width and height for each region
region_width = width // 3
region_height = height // 3

regions0 = []
regions1 = []
regions2 = []
regions3 = []
divided_regions = ["0000", "0010", "0110", "1000", "1010", "1110", "1001", "1011", "1111"]
items = np.array(pressure_df["Name"])
for item in items:
    part = item.split("_")
    x = int(part[3])
    y = int(part[4])

    row_index = y // region_height
    col_index = x // region_width

    region_number = row_index * 3 + col_index
    regions0.append(int(divided_regions[region_number][0]))
    regions1.append(int(divided_regions[region_number][1]))
    regions2.append(int(divided_regions[region_number][2]))
    regions3.append(int(divided_regions[region_number][3]))

new_columns = pd.DataFrame({
    225: regions0,
    226: regions1,
    227: regions2,
    228: regions3
})
pressure_df = pd.concat([pressure_df.iloc[:, :index_224 + 1], new_columns, pressure_df.iloc[:, index_224 + 1:]], axis=1)
```

```python
pressure_df
```

```python
insert_index = pressure_df.columns.get_loc(228) + 1
```

```python
pressure_df['Day'] = pressure_df['Day'].astype(int)
# Define a function to map day values to seasons
def map_season(day):
    year = (day - 1) // 365 + 1  # Calculate the year based on the day value
    day_within_year = (day - 1) % 365 + 1  # Calculate the day within the year

    if 1 <= day_within_year <= 90:          # Spring: Month 1-3
        return 1
    elif 91 <= day_within_year <= 180:      # Summer: Month 4-6
        return 2
    elif 181 <= day_within_year <= 270:     # Fall: Month 7-9
        return 3
    else:                                  # Winter: Month 10-12
        return 4

# Apply the function to the 'Day' column to create the 'season' column
pressure_df['season'] = pressure_df ['Day'].apply(map_season)
pressure_df.insert(insert_index, 'season', pressure_df.pop('season'))
```
