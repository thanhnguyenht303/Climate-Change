```python
import os
from PIL import Image
```

```python
image_folder = "/content/drive/MyDrive/10years_Merging_Image"
output_folder = "/content/drive/MyDrive/Double_Size_10years_Merging_Image"
# os.makedirs(output_folder)
```

```python
for file in os.listdir(image_folder):
    image_path = os.path.join(image_folder, file)
    image = Image.open(image_path)

    width, height = image.size
    n_width, n_height = width * 2, height * 2
    n_image = image.resize((n_width, n_height))

    n_image.save(os.path.join(output_folder, file))
```
