```python
import os
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Base_folder = "/content/drive/MyDrive/3rd_Wind_Pressure_vs_PressureOnly_Merging_Image"
Third_year_Pressure_Location_Season_SOM_Categorical_ROI_folder = "/content/drive/MyDrive/3rd_SOM_Categorical_Pressure_Adding_Location_Season_ROI_Image"
Third_year_Pressure_Location_Season_CNN_Categorical_ROI_folder = "/content/drive/MyDrive/3rd_CNN_Categorical_Pressure_Adding_Location_Season_ROI_Image"
Third_year_Wind_Pressure_Location_Season_SOM_Categorical_ROI_folder = "/content/drive/MyDrive/3rd_SOM_Categorical_Wind_Pressure_Adding_Location_Season_ROI_Image"
Third_year_Wind_Pressure_Location_Season_CNN_Categorical_ROI_folder = "/content/drive/MyDrive/3rd_CNN_Categorical_Wind_Pressure_Adding_Location_Season_ROI_Image"

Output_folder = "/content/drive/MyDrive/Merging_3rd_SOM_vs_CNN_Wind_Pressure_Location_Season_Image"
```

```python
# os.makedirs(Output_folder)
```

```python
for file in os.listdir(Base_folder):
    file_path1 = os.path.join(Base_folder, file)

    file_path2 = os.path.join(Third_year_Pressure_Location_Season_SOM_Categorical_ROI_folder, file)
    file_path3 = os.path.join(Third_year_Pressure_Location_Season_CNN_Categorical_ROI_folder, file)
    file_path4 = os.path.join(Third_year_Wind_Pressure_Location_Season_SOM_Categorical_ROI_folder, file)
    file_path5 = os.path.join(Third_year_Wind_Pressure_Location_Season_CNN_Categorical_ROI_folder, file)
    if os.path.exists(file_path2) and os.path.exists(file_path3) and os.path.exists(file_path4) and os.path.exists(file_path5):
        image1 = cv2.imread(file_path1)
        image2 = cv2.imread(file_path2)
        image3 = cv2.imread(file_path3)
        image4 = cv2.imread(file_path4)
        image5 = cv2.imread(file_path5)

        under_img = np.concatenate((image2, image3, image4, image5), axis = 1)

        height1, width1, _ = image1.shape
        height2, width2, _ = under_img.shape

        white_img = np.ones((height2, width1, 3), dtype = np.uint8) * 255
        start_x = 0

        white_img[:, start_x:start_x+width2, :] = under_img

        merged_img = np.concatenate((image1, white_img), axis = 0)

    output_file_path = os.path.join(Output_folder, file)
    cv2.imwrite(output_file_path, merged_img)
```
