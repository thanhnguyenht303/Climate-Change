```python
import math
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
pressure_first_2_year_path = "/content/drive/MyDrive/Location_Season_Data/Location_Season_Pressure_Training_ROI.csv"
pressure_wind_first_2_year_path = "/content/drive/MyDrive/Location_Season_Data/Location_Season_Wind_Pressure_Merging_Train_ROI.csv"

pressure_all_path = "/content/drive/MyDrive/Location_Season_Data/Location_Season_Pressure_ROI.csv"
pressure_wind_all_path = "/content/drive/MyDrive/Location_Season_Data/Location_Season_Pressure_Wind_Merging_ROI.csv"
```

```python
pressure_first_2_year_df = pd.read_csv(pressure_first_2_year_path)
pressure_first_2_year_df["Name"] = pressure_first_2_year_df["Name"].str.replace("Day", "")
pressure_first_2_year_df
```

```python
pressure_wind_first_2_year_df = pd.read_csv(pressure_wind_first_2_year_path)
pressure_wind_first_2_year_df["Name"] = pressure_wind_first_2_year_df["Name"].str.replace("Day", "")
pressure_first_2_year_df
```

```python
pressure_all_df = pd.read_csv(pressure_all_path)
pressure_all_df
```

```python
pressure_wind_all_df = pd.read_csv(pressure_wind_all_path)
pressure_wind_all_df
```

```python
pressure_3rd_year_df = pressure_all_df[(pressure_all_df["Day"] >= 730) & (pressure_all_df["Day"] <= 1095)]
pressure_3rd_year_df["ROI"] = pressure_3rd_year_df["Name"].apply(lambda x: x.split("_")[0] + "_" + x.split("_")[3] + "_" + x.split("_")[4])
pressure_3rd_year_df = pressure_3rd_year_df.drop_duplicates(subset = "ROI")
pressure_3rd_year_df = pressure_3rd_year_df[pressure_3rd_year_df["ROI"] != "738_73_20"]
pressure_3rd_year_df = pressure_3rd_year_df.sort_values(by = "Name")
pressure_3rd_year_df.reset_index(drop = True, inplace = True)
pressure_3rd_year_df
```

```python
third_year_labels = pd.read_table("/content/drive/MyDrive/Utils/3rd_year_Categorical_ROI.txt", sep = ":", header = None)
third_year_labels.rename(columns = {0: "Name", 1: "Class"}, inplace = True)
third_year_labels["Class"] = third_year_labels["Class"].str.strip()
third_year_labels = third_year_labels.sort_values(by = "Name")
third_year_labels.reset_index(drop = True, inplace = True)
third_year_labels["Class"] = third_year_labels["Class"].replace({'NROI': 3, 'COL': 2, 'COH': 1, 'CL': 0})
third_year_labels
```

```python
pressure_3rd_year_df["Class"] = third_year_labels["Class"]
pressure_3rd_year_df
```

```python
pressure_first_3_year_data = pd.concat([pressure_first_2_year_df, pressure_3rd_year_df.drop("ROI", axis = 1)], axis = 0)
# pressure_first_3_year_data.to_csv("/content/drive/MyDrive/Location_Season_Data/Location_Season_Pressure_first_3_year_ROI.csv", index = False)
pressure_first_3_year_data
```

```python
pressure_wind_3rd_year_df = pressure_wind_all_df[(pressure_wind_all_df["Day"] >= 730) & (pressure_wind_all_df["Day"] <= 1095)]
pressure_wind_3rd_year_df["ROI"] = pressure_wind_3rd_year_df["Name"].apply(lambda x: x.split("_")[0] + "_" + x.split("_")[3] + "_" + x.split("_")[4])
pressure_wind_3rd_year_df = pressure_wind_3rd_year_df.drop_duplicates(subset = "ROI")
pressure_wind_3rd_year_df = pressure_wind_3rd_year_df[pressure_wind_3rd_year_df["ROI"] != "738_73_20"]
pressure_wind_3rd_year_df = pressure_wind_3rd_year_df.sort_values(by = "Name")
pressure_wind_3rd_year_df.reset_index(drop = True, inplace = True)
pressure_wind_3rd_year_df["Class"] = third_year_labels["Class"]
pressure_wind_3rd_year_df
```

```python
pressure_wind_first_3_year_data = pd.concat([pressure_wind_first_2_year_df, pressure_wind_3rd_year_df.drop("ROI", axis = 1)], axis = 0)
pressure_wind_first_3_year_data.to_csv("/content/drive/MyDrive/Location_Season_Data/Location_Season_Wind_Pressure_first_3_year_ROI.csv", index = False)
pressure_wind_first_3_year_data
```
